```
import tabula
import pandas as pd
from google.colab import files

#Definir la ruta del PDF

pdf_ruta="https://github.com/josejparra/Diplomado-UNAL/raw/main/PreciosAgro2017.pdf"
```

```
#Corregir ceros

def chequeo(x):
  if not isinstance(x,str) and x<100:
    return x*1000
  else:
    return x
```

```
#Lectura del PDF usando ambos sistemas.

pag_l2 = tabula.read_pdf(pdf_ruta,pages="all",encoding="utf-8",pandas_options={'header': None},lattice=True)
pag_s2 = tabula.read_pdf(pdf_ruta,pages="all",encoding="utf-8",pandas_options={'header': None},stream=True)
```

```
#Nombres de los productos

nombres=["ACELGA","AGRAZ","AGUACATE","AGUACATE HASS",
         "p","AGUACATE HASS","AGUACATE LORENA","AHUYAMA","AJI",
         "p","AJO","APIO","ARRACACHA","ARROZ TECNIFICADO","ARROZ TRADICIONAL",
         "p","ARROZ TRADICIONAL","ARVEJA","BANANO CRIOLLO","BANANO MANZANO",
         "p","BANANO MONOCULTIVO","BOROJO","BREVO","BROCOLI","CACAO",
         "p","CACAO",
         "p","CAFE",
         "p","CAFE","CALABACÍN","PANELA",
         "p","PANELA",
         "p","PANELA","CARDAMOMO","CAUCHO","CEBOLLA DE HUEVO",
         "p","CEBOLLA JUNCA","CEBOLLIN","CHOCOLO","COCO",
         "p","CILANTRO","COL","COLIFLOR","CURCUMA","FIQUE",
         "p","ESPINACA","FRESA","FRIJOL ARBUSTIVO",
         "p","FRIJOL ARBUSTIVO","FRIJOL VERDE","FRIJOL VOLUBLE",
         "p","FRIJOL VOLUBLE","GRANADILLA",
         "p","GUANABANA","GUAYABA","GUAYABA MANZANA","GULUPA","HIGO",
         "p","HABICHUELA","HORTALIZAS VARIAS","LECHUGA","LIMA ACIDA TAHITI",
         "p","LIMON MANDARINO","LIMON PAJARITO","LULO",
         "p","MAIZ",
         "p","MAIZ","MAIZ FORRAJERO","MANDARINA CLEMENTINA",
         "p","MANDARINA ONECO","MANGO COMÚN","MANGO INJERTO","MARACUYA",
         "p","MARACUYA","MORA","MURRAPO","NARANJA TANGELO MINEOLA",
         "p","NARANJA COMÚN","NARANJA JAFFA","NARANJA TANGELO ORLANDO","NARANJA VALENCIA","NARANJA WASHINGTON NAVEL","NARANJA SALUSTIANA",
         "p","ÑAME","PALMA DE ACEITE","PAPA",
         "p","PAPA CRIOLLA","PAPAYA","PEPINO","PIMENTON","PIÑA",
         "p","PLANTAS AROMATICAS","PLATANO EN ASOCIO",
         "p","PLATANO EN ASOCIO","PLATANO MONOCULTIVO",
         "p","PLATANO MONOCULTIVO","REMOLACHA","REPOLLO","SACHA INCHI","STEVIA",
         "p","TOMATE CHONTO","TOMATE CHONTO INVERNADERO",
         "p","TOMATE CHONTO INVERNADERO","TOMATE DE ARBOL",
         "p","UCHUVA","YUCA",
         "p","YUCA",
         "p","ZANAHORIA"]

#Selección de lectura usando el sistema más apropiado

productos=pag_l2[1].applymap(chequeo)
productos["producto"]=nombres[0]
for i in range(2,len(pag_l2)):
    if pag_l2[i].iloc[0,0]=="Municipio":
      continue
    elif pag_l2[i].shape[1]==14 and pag_l2[i].iloc[0,0].find("\r")==-1:
        pag_l2[i]=pag_l2[i].applymap(chequeo)
        pag_l2[i]["producto"]=nombres[i-1]
        productos=productos.append(pag_l2[i])
    else:
      pag_s2[i]=pag_s2[i].applymap(chequeo)
      pag_s2[i]["producto"]=nombres[i-1]
      productos=productos.append(pag_s2[i])
```

```
#Añadir columnas que contienen nombres
columnas=pag_l2[0].loc[0,:].tolist()
columnas.append("producto")
productos.columns=columnas

#Verificar que no haya ningún producto con el cambio de página.
productos[productos.producto=="p"]
```

```
#Añadir código municipal del DANE.

divipola_ruta="https://raw.githubusercontent.com/josejparra/Diplomado-UNAL/main/Divipola.csv"

divipola=pd.read_csv(divipola_ruta,encoding="latin-1",sep=";")
divipola.rename(columns={"Municipios":"Municipio"},inplace=True)

divipola.Municipio=divipola.Municipio.replace("Jerico","Jericó")
divipola.Municipio=divipola.Municipio.replace("Cañasgordas","Cañas Gordas")

#Verificar que todos los municipios tengan nombres homogéneos.
divipola.Municipio.unique()
```

```
#Unificar nombres.

productos.Municipio=productos.Municipio.replace("Cañasgordas","Cañas Gordas")
productos2=productos.merge(divipola,"left")

productos2
```

```
#Eliminar promedios

productos3=productos2[~(productos2.Municipio=="Promedio")]

productos3
```

```
productos3[productos3.Codigo.isnull()]

productos3.set_index("Codigo",inplace=True)
```
